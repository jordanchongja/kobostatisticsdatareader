```python
import os
import re
import yaml
import requests
import shutil
from pathlib import Path
from datetime import datetime

# --- Configuration ---
OBSIDIAN_DIR = r"C:\Obsidian\🧠 Media and Growth\Books" 
QUARTO_BASE_DIR = r"C:\Github Code\jordanchongja.github.io\readings"

# Set to True to force update the index.qmd files. 
FORCE_OVERWRITE_QMD = False

def convert_rating(rating_str):
    """Converts a numeric string rating to star symbols."""
    try:
        rating = int(rating_str)
        return "★" * rating
    except (ValueError, TypeError):
        return ""

def download_image(url, save_path):
    """Downloads the cover image with fallback for missing high-res versions."""
    if not url:
        return False
    
    headers = {'User-Agent': 'Mozilla/5.0'}
    urls_to_try = [url]
    
    # If it's a Google Books URL, prep a high-res attempt and a fallback attempt
    if "google.com" in url:
        high_res_url = url.replace("zoom=1", "zoom=3").replace("&edge=curl", "")
        urls_to_try = [high_res_url, url]
    
    for try_url in urls_to_try:
        try:
            response = requests.get(try_url, headers=headers, timeout=10)
            if response.status_code == 200:
                # The generic Google Books placeholder is typically under 3KB.
                # If the image is > 4000 bytes, it's a real cover.
                # Alternatively, if it's our last resort URL, we save it anyway.
                if len(response.content) > 4000 or try_url == urls_to_try[-1]:
                    with open(save_path, 'wb') as out_file:
                        out_file.write(response.content)
                    return True
        except Exception as e:
            print(f"Failed on {try_url}: {e}")
            
    return False

def clean_body_content(body):
    """Removes Obsidian syntax and fixes Quarto YAML parsing errors."""
    body = re.sub(r'%%.*?%%', '', body, flags=re.DOTALL)
    body = re.sub(r'!\[cover\|\d+\].*?\)', '', body)
    body = re.sub(r'^---$', '***', body, flags=re.MULTILINE)
    return body.strip()

def process_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        content = f.read()

    match = re.match(r'^---\n(.*?\n)---\n(.*)', content, re.DOTALL)
    if not match:
        print(f"Skipping {file_path.name}: No valid frontmatter found.")
        return

    frontmatter_str, body_content = match.groups()
    
    try:
        meta = yaml.safe_load(frontmatter_str)
    except yaml.YAMLError as e:
        print(f"Error parsing YAML in {file_path.name}: {e}")
        return

    # --- EXTRACT OBSIDIAN METADATA ---
    title = meta.get('title', file_path.stem)
    
    # Handle author if it's a list (like in your Obsidian format)
    author_val = meta.get('author', 'Unknown')
    author = author_val[0] if isinstance(author_val, list) and author_val else author_val
    if isinstance(author, list): # Fallback if empty list
        author = "Unknown"

    # Handle categories
    obsidian_categories = meta.get('category', [])
    categories = obsidian_categories if isinstance(obsidian_categories, list) else [obsidian_categories]
    if 'Book' not in categories:
        categories.append('Book')

    # Handle Date (Created timestamp -> YYYY-MM-DD)
    date_val = meta.get('created', meta.get('publish', datetime.now()))
    if isinstance(date_val, datetime):
        date_str = date_val.strftime('%Y-%m-%d')
    else:
        date_str = str(date_val).split(' ')[0] 

    # --- PATH SETUP ---
    safe_title = re.sub(r'[\\/*?:"<>|]', "", title)
    book_dir = Path(QUARTO_BASE_DIR) / safe_title
    output_qmd_path = book_dir / "index.qmd"
    cover_path = book_dir / "cover.jpg"

    # --- OVERWRITE PROTECTION FOR QMD ---
    if output_qmd_path.exists() and not FORCE_OVERWRITE_QMD:
        print(f"--- Skipping {title}: File already exists.")
        return
    
    book_dir.mkdir(parents=True, exist_ok=True)

    # --- RATING & SUBTITLE COMBINATION ---
    raw_rating = meta.get('rating', '')
    rating_stars = convert_rating(raw_rating)
    obsidian_subtitle = meta.get('subtitle', '')
    
    # Combine stars and subtitle nicely for the grid UI
    if rating_stars and obsidian_subtitle:
        final_subtitle = f"{rating_stars} | {obsidian_subtitle}"
    elif rating_stars:
        final_subtitle = rating_stars
    elif obsidian_subtitle:
        final_subtitle = obsidian_subtitle
    else:
        final_subtitle = ""

    numeric_rating = 0
    try:
        if raw_rating:
            numeric_rating = int(raw_rating)
    except (ValueError, TypeError):
        pass

    # --- IMAGE OVERWRITE PROTECTION & DOWNLOAD ---
    image_url = meta.get('cover', '')
    if image_url and not cover_path.exists():
        download_image(image_url, cover_path)

    # --- PREPARE QUARTO FRONTMATTER ---
    quarto_frontmatter = {
        'title': title,
        'author': author,
        'date': date_str,
        'categories': categories,
        'image': "cover.jpg",
        'subtitle': final_subtitle,
        'rating': numeric_rating, # Ensure this is ALWAYS added now
        'body-class': "project-detail-view",
        'include-before-body': {
            'text': "<div style='margin-top: 50px;'></div>"
        }
    }
    
    if numeric_rating is not None:
        quarto_frontmatter['rating'] = numeric_rating

    # --- WRITE FILE ---
    cleaned_body = clean_body_content(body_content)
    css_block = """
<style>
  #quarto-margin-sidebar {
    padding-top: 50px;
  }
</style>

"""
    with open(output_qmd_path, 'w', encoding='utf-8') as f:
        f.write("---\n")
        # allow_unicode=True keeps the stars from turning into escape codes
        yaml.dump(quarto_frontmatter, f, default_flow_style=False, sort_keys=False, allow_unicode=True)
        f.write("---\n\n")
        f.write(css_block)
        f.write(cleaned_body)

    action = "UPDATED" if output_qmd_path.exists() and FORCE_OVERWRITE_QMD else "NEW"
    print(f"Successfully processed {action} book: {title}")

def main():
    obsidian_path = Path(OBSIDIAN_DIR)
    if not obsidian_path.exists():
        print(f"Directory not found: {OBSIDIAN_DIR}")
        return

    for md_file in obsidian_path.glob("*.md"):
        process_file(md_file)

if __name__ == "__main__":
    main()
```
