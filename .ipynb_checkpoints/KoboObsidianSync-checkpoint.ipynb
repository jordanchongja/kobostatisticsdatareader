```python
# Standard Library Imports
import glob
import io
import json
import os
import re
import sqlite3
import struct
import urllib.parse
from collections import defaultdict
from datetime import datetime
from difflib import SequenceMatcher

# Third-Party Library Imports
import pandas as pd

# 1. Define your file paths
db_path = "Kobo\KoboReader.sqlite"
DB_PATH = "Kobo\KoboReader.sqlite"

# --- 1. DEFINE YOUR FILE PATHS ---
output_dir = r"output"
obsidian_dir = r"C:\Obsidian\🧠 Media and Growth\Books"
to_add_file_path = os.path.join(obsidian_dir, "00_Books to Add.md")
wordlist_md_path = os.path.join(obsidian_dir, "00_wordlist.md")
```

```python
import os
import io
import json
import sqlite3
import struct
from datetime import datetime
import pandas as pd

# --- 1. UTILITY CLASSES & FUNCTIONS ---
class KoboBinaryReader:
    def __init__(self, data_bytes):
        self.data = data_bytes

    def extract_timestamps(self):
        target = "eventTimestamps".encode('utf-16-be')
        idx = self.data.find(target)
        if idx == -1: return []
        stream = io.BytesIO(self.data[idx + len(target):])
        try:
            v_type = struct.unpack(">I", stream.read(4))[0]
            is_null = struct.unpack(">?", stream.read(1))[0]
            if v_type == 9 and not is_null: 
                count = struct.unpack(">I", stream.read(4))[0]
                timestamps = []
                for _ in range(count or 0):
                    i_type = struct.unpack(">I", stream.read(4))[0]
                    i_null = struct.unpack(">?", stream.read(1))[0]
                    if i_type == 3: 
                        timestamps.append(struct.unpack(">I", stream.read(4))[0])
                    else: stream.read(4)
                return timestamps
        except: pass
        return []

def cluster_and_scale(ts_list, target_total_sec, gap_limit=300):
    if not ts_list: return [], 1.0
    sorted_ts = sorted(list(set(ts_list)))
    
    raw_sessions = []
    if not sorted_ts: return [], 1.0
    
    start_ts = sorted_ts[0]
    last_ts = sorted_ts[0]
    pages = 1
    for i in range(1, len(sorted_ts)):
        if sorted_ts[i] - last_ts < gap_limit:
            pages += 1
        else:
            raw_sessions.append({'start': start_ts, 'end': last_ts, 'pages': pages})
            start_ts = sorted_ts[i]
            pages = 1
        last_ts = sorted_ts[i]
    raw_sessions.append({'start': start_ts, 'end': last_ts, 'pages': pages})
    
    raw_total_duration = sum([(s['end'] - s['start'] + 60) for s in raw_sessions])
    
    # --- THE ANTI-ZERO FIX ---
    # If Kobo's clock is broken (target_total_sec is smaller than our raw calculated time),
    # we completely abandon Kobo's clock and trust the raw page-turn timestamps.
    if raw_total_duration > 0 and target_total_sec < raw_total_duration:
        scaling_factor = 1.0 
    else:
        scaling_factor = target_total_sec / raw_total_duration if raw_total_duration > 0 else 1.0
    
    final_sessions = []
    for s in raw_sessions:
        corrected_dur_sec = ((s['end'] - s['start']) + 60) * scaling_factor
        dur_mins = round(corrected_dur_sec / 60, 2)
        
        # HARD LIMIT: A session can never be 0 minutes. 
        # If you turned a page, you spent at least 1 minute.
        if dur_mins < 1.0:
            dur_mins = 1.0
            
        final_sessions.append({
            'timestamp': datetime.fromtimestamp(s['start']).isoformat(),
            'duration_minutes': dur_mins,
            'pages_read': s['pages']
        })
    return final_sessions, scaling_factor

def safe_decode(val, default_val="Unknown"):
    """Safely handles NoneTypes, Strings, and Bytes during extraction."""
    if pd.isnull(val): 
        return default_val
    if isinstance(val, bytes): 
        return val.decode('utf-8', errors='ignore')
    return str(val)

# --- 2. MAIN CONSOLIDATED EXPORTER ---
def export_kobo_full_system(db_path, output_dir="output"):
    if not os.path.exists(db_path):
        print(f"❌ ERROR: Database not found at {db_path}")
        return

    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    save_path = os.path.join(output_dir, f"kobo_consolidated_{timestamp}.json")

    conn = sqlite3.connect(db_path)
    conn.text_factory = bytes 
    cursor = conn.cursor()

    # --- PART A: MASTER READING ANALYSIS ---
    
    # Dynamic Column Detection
    cursor.execute("PRAGMA table_info(content);")
    all_cols = [col[1].decode('utf-8') if isinstance(col[1], bytes) else col[1] for col in cursor.fetchall()]
    percent_col = "___PercentRead" if "___PercentRead" in all_cols else "PercentRead"

    # 1. Fetch Narrative Metadata (WITH LEFT JOIN SAFETY NET)
    df_metadata = pd.read_sql_query(f"""
        SELECT 
            Parent.ContentID, Parent.Title, Parent.Attribution as Author, 
            Parent.TimeSpentReading, Parent.{percent_col} as Percent,
            Parent.ReadStatus, 
            SUM(CASE WHEN Children.ContentType = 9 AND Children.WordCount > 0 THEN Children.WordCount ELSE 0 END) as NarrativeWordCount
        FROM content AS Parent
        LEFT JOIN content AS Children ON Children.BookID = Parent.ContentID
        WHERE Parent.ContentType = 6 AND Parent.TimeSpentReading > 0
        GROUP BY Parent.ContentID
    """, conn)

    # 2. Fetch Events & Highlights
    golden_codes = [3, 5, 7, 8, 9, 46, 80, 1012, 1013, 1014]
    df_events = pd.read_sql_query(f"SELECT ContentID, ExtraData FROM Event WHERE EventType IN ({','.join(map(str, golden_codes))})", conn)
    df_bookmarks = pd.read_sql_query("SELECT VolumeID, Text, DateCreated FROM Bookmark WHERE Text IS NOT NULL", conn)

    processed_reading_analysis = {}
    
    print(f"--- 🚀 EXTRACTING ANALYTICS FOR {len(df_metadata)} BOOKS ---")
    
    for _, row in df_metadata.iterrows():
        # SAFE DECODE IMPLEMENTATION
        cid = safe_decode(row['ContentID'], "Unknown_ID")
        title = safe_decode(row['Title'], "Unknown Title")
        author = safe_decode(row['Author'], "Unknown Author")
        
        system_sec = row['TimeSpentReading']
        
        # Extract Sessions
        book_ts = []
        relevant_events = df_events[df_events['ContentID'].apply(lambda x: safe_decode(x, "") == cid or title in safe_decode(x, ""))]
        for blob in relevant_events['ExtraData']:
            book_ts.extend(KoboBinaryReader(blob).extract_timestamps())
            
        sessions, factor = cluster_and_scale(book_ts, system_sec)
        
        total_min = system_sec / 60
        
        # ReadStatus Override
        raw_percent = row['Percent'] if pd.notnull(row['Percent']) else 0
        read_status = row['ReadStatus'] if pd.notnull(row['ReadStatus']) else 0
        
        if read_status == 2:
            safe_percent = 100  
        else:
            safe_percent = max(0, min(100, raw_percent))
            
        wpm = round(((row['NarrativeWordCount'] or 0) * (safe_percent / 100.0)) / total_min, 1) if total_min > 0 else 0
        
        # Safe Decode for Highlights
        hls = [{"text": safe_decode(h['Text'], ""), "date": safe_decode(h['DateCreated'], "")} 
               for _, h in df_bookmarks[df_bookmarks['VolumeID'].apply(lambda x: safe_decode(x, "") == cid or title in safe_decode(x, ""))].iterrows()]

        processed_reading_analysis[cid] = {
            "metadata": {"title": title, "author": author, "word_count": row['NarrativeWordCount'], "percent_complete": safe_percent},
            "metrics": {"total_minutes": round(total_min, 1), "avg_wpm": wpm, "session_count": len(sessions), "scaling_factor": round(factor, 3)},
            "sessions": sessions,
            "highlights": hls
        }

    # --- PART B: RAW DATABASE DUMP ---
    print("--- 📦 BACKING UP RAW DATABASE TABLES ---")
    cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
    tables = [t[0].decode('utf-8') for t in cursor.fetchall()]
    raw_database_dump = {}

    for table in tables:
        try:
            df_table = pd.read_sql_query(f"SELECT * FROM {table}", conn)
            for col in df_table.columns:
                if df_table[col].dtype == object:
                    df_table[col] = df_table[col].apply(lambda x: x.hex() if isinstance(x, bytes) else x)
            raw_database_dump[table] = df_table.to_dict(orient='records')
        except:
            continue

    conn.close()

    # --- PART C: FINAL CONSOLIDATION ---
    final_output = {
        "sync_info": {
            "generated_at": datetime.now().isoformat(),
            "device_source": "Kobo Rig Consolidated"
        },
        "reading_dashboard": processed_reading_analysis,
        "raw_database_backup": raw_database_dump
    }

    with open(save_path, 'w', encoding='utf-8') as f:
        json.dump(final_output, f, indent=4, ensure_ascii=False)

    print(f"\n✨ SUCCESS: Consolidated Reading Analysis and Full DB Dump saved to: {save_path}")

export_kobo_full_system(DB_PATH)
```

```python
import os
import json
import glob
import re
import urllib.parse
from difflib import SequenceMatcher

# --- 2. CUSTOM BOOK MAP ---
# Map the messy Kobo Title exactly to your clean Obsidian Title (the part before the hyphen)
CUSTOM_BOOK_MAP = {
    'East of Eden (Steinbeck "Essentials")': 'East of Eden',
    # Add more here as needed:
    # 'Messy Publisher Title': 'Clean Title',
}

# --- UTILITY FUNCTIONS ---
def fuzzy_match(string1, string2, threshold=0.8):
    s1 = "".join(c.lower() for c in string1 if c.isalnum())
    s2 = "".join(c.lower() for c in string2 if c.isalnum())
    return SequenceMatcher(None, s1, s2).ratio() >= threshold

def inject_or_update_yaml(content, wpm, word_count, percent):
    yaml_pattern = re.compile(r"^---\n(.*?)\n---", re.DOTALL)
    new_stats = f"wpm: {wpm}\nword_count: {word_count}\npercent_complete: {percent}"
    if content.startswith("---"):
        def replace_or_add(match):
            yaml_block = match.group(1)
            yaml_block = re.sub(r"^(wpm|word_count|percent_complete):.*\n?", "", yaml_block, flags=re.MULTILINE)
            return f"---\n{yaml_block.strip()}\n{new_stats}\n---"
        return yaml_pattern.sub(replace_or_add, content, count=1)
    else:
        return f"---\n{new_stats}\n---\n\n{content}"

def decode_kobo_text(text):
    if not isinstance(text, str) or not text: return ""
    try:
        return bytes.fromhex(text).decode('utf-8')
    except (ValueError, UnicodeDecodeError): return text 

def parse_volume_id(volume_id):
    if not volume_id: return "Unknown Book"
    clean_path = urllib.parse.unquote(volume_id)
    filename = clean_path.split('/')[-1]
    filename = filename.replace('.kepub.epub', '').replace('.epub', '')
    title = filename.split(' - ')[0]
    return title.strip()

# --- MAIN SYNC ENGINE ---
def sync_kobo_to_obsidian():
    list_of_files = glob.glob(os.path.join(output_dir, "kobo_consolidated_*.json"))
    if not list_of_files:
        print("❌ ERROR: No Kobo JSON files found in the output directory.")
        return
    latest_json = max(list_of_files, key=os.path.getctime)
    
    with open(latest_json, 'r', encoding='utf-8') as f:
        master_data = json.load(f)
        
    books_data = master_data.get("reading_dashboard", master_data.get("books", {}))
    raw_db = master_data.get("raw_database_backup", {})
    
    md_files = glob.glob(os.path.join(obsidian_dir, "*.md"))
    vault_map = {}
    for fp in md_files:
        filename = os.path.basename(fp).replace(".md", "")
        if filename in ["00_Books to Add", "00_wordlist"]: continue
        vault_title = filename.split(" - ")[0].strip()
        vault_map[vault_title] = fp

    print(f"--- 🚀 INITIATING OBSIDIAN SYNC ---")
    print(f"📊 Found {len(books_data)} books in Kobo database.")
    print(f"📂 Found {len(vault_map)} existing book notes in Vault.\n")

    if os.path.exists(to_add_file_path):
        with open(to_add_file_path, 'r', encoding='utf-8') as f:
            to_add_content = f.read()
    else:
        to_add_content = "# Books to Add\n\n"

    for cid, book_info in books_data.items():
        raw_kobo_title = book_info['metadata']['title']
        kobo_author = book_info['metadata'].get('author', 'Unknown Author')
        
        # --- THE MAPPING OVERRIDE ---
        # If the title is in our map, translate it. Otherwise, use the original.
        search_title = CUSTOM_BOOK_MAP.get(raw_kobo_title, raw_kobo_title)
        
        matched_path = None
        
        # 1. Try Exact Match First (Instant link if mapped)
        if search_title in vault_map:
            matched_path = vault_map[search_title]
        # 2. Fallback to Fuzzy Match
        else:
            for vault_title, fp in vault_map.items():
                if fuzzy_match(vault_title, search_title):
                    matched_path = fp
                    break
        
        if not matched_path:
            task_string = f"- [ ] {raw_kobo_title} - {kobo_author}"
            if task_string not in to_add_content:
                to_add_content += f"{task_string}\n"
                print(f"📥 Added to Inbox: {raw_kobo_title}")
            continue

        print(f"✅ Syncing Note: {os.path.basename(matched_path)}")
        
        with open(matched_path, 'r', encoding='utf-8') as f:
            content = f.read()

        metrics = book_info.get('quant_metrics', book_info.get('metrics', {}))
        metadata = book_info['metadata']
        content = inject_or_update_yaml(
            content, wpm=metrics.get('avg_wpm', 0), word_count=metadata.get('word_count', 0), percent=metadata.get('percent_complete', 0)
        )

        journey_lines = ["## Reading Journey"]
        total_time_float = 0.0 
        cumulative_pages = 0
        
        sessions = book_info.get('sessions', [])
        
        if sessions:
            first_date = sessions[0]['timestamp'][:10]
            journey_lines.append(f"- {first_date}: Started")
        
        for session in sessions:
            date_str = session['timestamp'][:10]
            pages = session['pages_read']
            duration_float = session.get('duration_minutes', 0.0)
            
            cumulative_pages += pages
            total_time_float += duration_float
            
            duration_display = max(1, round(duration_float))
            
            journey_lines.append(f"- {date_str}: {cumulative_pages}")
            journey_lines.append(f"    - {duration_display} mins")
            
        if metadata.get('percent_complete', 0) > 95:
            last_date = sessions[-1]['timestamp'][:10] if sessions else "Unknown Date"
            journey_lines.append(f"- {last_date}: Finished")
            
        hours, mins = divmod(int(total_time_float), 60)
        if hours > 0:
            time_formatted = f"{hours} hours and {mins} minutes"
        else:
            time_formatted = f"{mins} minutes"
            
        journey_lines.append(f"- Total Time Taken: {time_formatted}")
        journey_str = "\n".join(journey_lines)

        if "## Reading Journey" in content:
            content = re.sub(r"## Reading Journey.*?(?=\n## |\Z)", journey_str, content, flags=re.DOTALL)
        else:
            content += f"\n\n{journey_str}"

        if "## Bookmarks" not in content:
            content += "\n\n## Bookmarks\n"

        bookmarks_to_add = []
        for hl in book_info['highlights']:
            clean_text = hl['text'].strip()
            if clean_text and clean_text not in content:
                date_str = hl['date'][:10] if hl['date'] else ""
                bookmarks_to_add.append(f"> {clean_text}\n> *{date_str}*\n")

        if bookmarks_to_add:
            content = content.replace("## Bookmarks", "## Bookmarks\n" + "\n".join(bookmarks_to_add))

        with open(matched_path, 'w', encoding='utf-8') as f:
            f.write(content)

    with open(to_add_file_path, 'w', encoding='utf-8') as f:
        f.write(to_add_content)

    # Wordlist sync
    kobo_words = raw_db.get("WordList", [])
    if kobo_words:
        print(f"\n--- 📖 SYNCING VOCABULARY ({len(kobo_words)} words in database) ---")
        existing_words = set()
        if os.path.exists(wordlist_md_path):
            with open(wordlist_md_path, 'r', encoding='utf-8') as f:
                wl_content = f.read()
                existing_words = set(re.findall(r"^###\s+(.*)", wl_content, re.MULTILINE))
        else:
            with open(wordlist_md_path, 'w', encoding='utf-8') as f:
                f.write("# 🧠 Vocabulary Inbox\n\n")
                
        new_entries = []
        for item in kobo_words:
            raw_word = item.get('Text', item.get('Word', '')).strip()
            word = decode_kobo_text(raw_word)
            raw_def = item.get('DictString', item.get('Definition', '')).strip()
            definition = decode_kobo_text(raw_def).replace("\n", " ")
            raw_ctx = item.get('ContextString', item.get('Snippet', '')).strip()
            context = decode_kobo_text(raw_ctx).replace("\n", " ")
            raw_date = item.get('DateCreated', '')
            date_str = decode_kobo_text(raw_date)[:10] if raw_date else "Unknown Date"
            raw_vol = item.get('VolumeId', item.get('VolumeID', ''))
            
            # Apply the mapping to the vocabulary sources too!
            book_title = parse_volume_id(decode_kobo_text(raw_vol))
            book_title = CUSTOM_BOOK_MAP.get(book_title, book_title)
            
            if not word or word in existing_words: continue
                
            entry = f"### {word}\n"
            entry += f"> **Source:** {book_title} (*{date_str}*)\n"
            if definition: entry += f"> **Definition:** {definition}\n"
            if context: entry += f"> *\"{context}\"*\n"
            entry += "\n"
            
            new_entries.append(entry)
            existing_words.add(word) 

        if new_entries:
            with open(wordlist_md_path, 'a', encoding='utf-8') as f:
                f.write("".join(new_entries))
            print(f"✅ Added {len(new_entries)} new words to 00_wordlist.md")
        else:
            print("🤷 No new words to add.")

    print(f"\n✨ SUCCESS: Obsidian Sync Complete!")

if __name__ == "__main__":
    sync_kobo_to_obsidian()
```

```python
import sqlite3
import pandas as pd
import urllib.parse
import struct
import io
from datetime import datetime

class KoboBinaryReader:
    def __init__(self, data_bytes):
        self.data = data_bytes

    def extract_timestamps(self):
        target = "eventTimestamps".encode('utf-16-be')
        idx = self.data.find(target)
        if idx == -1: return []
        stream = io.BytesIO(self.data[idx + len(target):])
        try:
            v_type = struct.unpack(">I", stream.read(4))[0]
            is_null = struct.unpack(">?", stream.read(1))[0]
            if v_type == 9 and not is_null: 
                count = struct.unpack(">I", stream.read(4))[0]
                timestamps = []
                for _ in range(count or 0):
                    i_type = struct.unpack(">I", stream.read(4))[0]
                    i_null = struct.unpack(">?", stream.read(1))[0]
                    if i_type == 3: 
                        timestamps.append(struct.unpack(">I", stream.read(4))[0])
                    else: stream.read(4)
                return timestamps
        except: pass
        return []

def cluster_and_scale(ts_list, target_total_sec, gap_limit=300):
    # FIX: Return only an empty list to keep types consistent
    if not ts_list: 
        return [] 
        
    sorted_ts = sorted(list(set(ts_list)))
    if not sorted_ts: 
        return []

    raw_sessions = []
    start_ts = sorted_ts[0]
    last_ts = sorted_ts[0]
    pages = 1

    for i in range(1, len(sorted_ts)):
        if sorted_ts[i] - last_ts < gap_limit:
            pages += 1
        else:
            raw_sessions.append({'start': start_ts, 'end': last_ts, 'pages': pages})
            start_ts = sorted_ts[i]
            pages = 1
        last_ts = sorted_ts[i]
    raw_sessions.append({'start': start_ts, 'end': last_ts, 'pages': pages})
    
    raw_total_duration = sum([(s['end'] - s['start'] + 60) for s in raw_sessions])
    
    # Logic to scale reconstructed time to match Kobo's recorded TimeSpentReading
    if raw_total_duration > 0 and target_total_sec < raw_total_duration:
        scaling_factor = 1.0 
    else:
        scaling_factor = target_total_sec / raw_total_duration if raw_total_duration > 0 else 1.0
    
    final_sessions = []
    for s in raw_sessions:
        corrected_dur_sec = ((s['end'] - s['start']) + 60) * scaling_factor
        dur_mins = max(1.0, round(corrected_dur_sec / 60, 2))
        final_sessions.append({'duration_minutes': dur_mins})
        
    return final_sessions # Always returns a list of dicts

def decode_kobo_text(text):
    if not isinstance(text, str) or not text: return ""
    try:
        if all(c in "0123456789abcdefABCDEF" for c in text) and len(text) > 2:
            return bytes.fromhex(text).decode('utf-8')
    except: pass
    return text 

def run_diagnostics():
    conn = sqlite3.connect(DB_PATH)
    conn.text_factory = bytes
    
    print("==================================================")
    print(" 🕰️ TIME DISCREPANCY AUDIT")
    print("==================================================")
    
    # 1. Get raw metadata time
    df_content = pd.read_sql_query("SELECT ContentID, Title, TimeSpentReading FROM content WHERE ContentType = 6", conn)
    raw_metadata_sec = 0
    for val in df_content['TimeSpentReading']:
        if pd.notnull(val): raw_metadata_sec += val
    
    print(f"Total Time according to Kobo Metadata (TimeSpentReading): {raw_metadata_sec / 3600:.2f} hours")
    
    # 2. Get session reconstructed time
    df_events = pd.read_sql_query("SELECT ContentID, ExtraData FROM Event WHERE EventType IN (3,5,7,8,9,46,80,1012,1013,1014)", conn)
    
    total_session_mins = 0
    for _, row in df_content.iterrows():
        cid = row['ContentID'].decode('utf-8', 'ignore') if isinstance(row['ContentID'], bytes) else str(row['ContentID'])
        system_sec = row['TimeSpentReading'] if pd.notnull(row['TimeSpentReading']) else 0
        
        book_ts = []
        rel_events = df_events[df_events['ContentID'].apply(lambda x: (x.decode('utf-8', 'ignore') if isinstance(x, bytes) else str(x)) == cid)]
        for blob in rel_events['ExtraData']:
            book_ts.extend(KoboBinaryReader(blob).extract_timestamps())
            
        sessions = cluster_and_scale(book_ts, system_sec)
        
        # Add fallback just like our app does
        if not sessions and system_sec > 0:
            sessions = [{'duration_minutes': round(system_sec/60, 2)}]
            
        for s in sessions:
            total_session_mins += s['duration_minutes']
            
    print(f"Total Time according to Reconstructed Sessions:       {total_session_mins / 60:.2f} hours")
    print(f"Difference:                                           {abs((raw_metadata_sec / 3600) - (total_session_mins / 60)):.2f} hours")
    print("\n")


    print("==================================================")
    print(" 📖 VOCABULARY MAPPING AUDIT ('The Vegetarian')")
    print("==================================================")
    
    df_words = pd.read_sql_query("SELECT * FROM WordList", conn)
    
    # Create a quick map for lookup checking
    content_map = {}
    for cid, title in zip(df_content['ContentID'], df_content['Title']):
        c_str = cid.decode('utf-8', 'ignore') if isinstance(cid, bytes) else str(cid)
        t_str = title.decode('utf-8', 'ignore') if isinstance(title, bytes) else str(title)
        content_map[c_str] = t_str

    # Find the problematic words
    for _, row in df_words.iterrows():
        w_text_raw = row.get('Text', b'').decode('utf-8', 'ignore') if isinstance(row.get('Text'), bytes) else str(row.get('Text', ''))
        word = decode_kobo_text(w_text_raw)
        
        # Let's specifically look for words from The Vegetarian
        if word in ["taciturn", "engagé", "ascetic", "countenance"]:
            w_vol_raw = row.get('VolumeId', row.get('VolumeID', b'')).decode('utf-8', 'ignore') if isinstance(row.get('VolumeId'), bytes) else str(row.get('VolumeId'))
            w_vol_decoded = decode_kobo_text(w_vol_raw)
            w_vol_norm = urllib.parse.unquote(w_vol_decoded).lower()
            
            print(f"\n--- Testing word: {word} ---")
            print(f"1. Raw Hex from WordList:  {w_vol_raw}")
            print(f"2. Decoded VolumeId:       {w_vol_decoded}")
            print(f"3. Unquoted & Lowercased:  {w_vol_norm}")
            
            # Check matches
            match_found = False
            for cid_key, title_val in content_map.items():
                norm_cid = urllib.parse.unquote(cid_key).lower()
                if norm_cid in w_vol_norm or w_vol_norm in norm_cid:
                    print(f"✅ MATCH SUCCESS! Linked to Content Table Title: '{title_val}'")
                    match_found = True
                    break
            
            if not match_found:
                print("❌ MATCH FAILED! Here is what the library has for 'The Vegetarian':")
                for cid_key, title_val in content_map.items():
                    if "vegetarian" in title_val.lower():
                        print(f"   ContentID in Library: {cid_key}")
                        print(f"   Title in Library:     {title_val}")

    conn.close()

run_diagnostics()
```
